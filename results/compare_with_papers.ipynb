```python
#collect RA and dec of all images
import pandas as pd
import os
import time 
import subprocess
from astropy import wcs
from astropy.io import fits
from astropy.nddata.utils import Cutout2D
import numpy as np
import random
import math
from astropy import units as u
from astropy.coordinates import SkyCoord

precision =0.1#.5 #try 0.25
thresh = 0.3
PATH = '/home/roberttoyonaga/notebooks/roberttoyonaga/CMUDeepLensNRC/HSTLens/results/'#analysis/results_1/'
CSV = 'paper_crds.csv'
papers = pd.read_csv(CSV, sep = ',')  #ra, dec, author
matches = {'paper_ra':[],'paper_dec':[],'det_ra':[],'det_dec':[], 'prb_1':[], 'obj_name':[], 'author':[]}

def get_crds(filename):
    crd_dict = {"ra": [], "dec":[], "obj_name":[], 'prb_1':[] } 
    df = pd.read_csv(filename, sep = ',') #open results csv
    for row in range(len(df)): #for each row
        if float(df['prb_1'][row]) > thresh:#if lens prb >thresh

            crd_dict['ra'].append(float(df['ra'][row]))
            crd_dict['dec'].append(float(df['dec'][row]))
            crd_dict['obj_name'].append(df['obj_name'][row])
            crd_dict['prb_1'].append(float(df['prb_1'][row]))

    return crd_dict

def compare(crd_dict): #compare each of the crd sets in the papers to the current results file 
    for i in range(len(papers['ra'])):
        if (":" in str(papers['ra'][i])) or(" " in str(papers['ra'][i])):
            try:
                r = papers['ra'][i]
                d= papers['dec'][i]
                crd = SkyCoord(r[0:2]+"h"+r[3:5]+'m'+r[6:10]+'s',d[0:3]+"d"+d[4:6]+'m'+d[7:9]+'s')
                papers['ra'][i] = crd.ra.degree
                papers['dec'][i] = crd.dec.degree
            except:
                print('unable to convert crds')
                continue
            
        for j in range(len(crd_dict['ra'])):
#             try:
                d1= float(papers['dec'][i])*0.0174533
                d2= float(crd_dict['dec'][j])*0.0174533
                r1= float(papers['ra'][i])*0.0174533
                r2= float(crd_dict['ra'][j])*0.0174533
                theta = (np.arccos(np.sin(d1)*np.sin(d2)+np.cos(d1)*np.cos(d2)*np.cos(r1-r2)))*3600 #multiply by 3600 to get arcseconds
                
                if theta < precision:
                    matches['paper_ra'].append(papers['ra'][i])
                    matches['paper_dec'].append(papers['dec'][i])
                    matches['det_ra'].append(crd_dict['ra'][j])
                    matches['det_dec'].append(crd_dict['dec'][j])
                    matches['prb_1'].append(crd_dict['prb_1'][j])
                    matches['obj_name'].append(crd_dict['obj_name'][j])
                    matches['author'].append(papers['author'][i])
                    break
                
#             except:
#                 print('failed')
#                 continue
                
    
        
        
count=0 
        
for filename in os.listdir(PATH):
    count+=1
    if count==1000:
        break
    print(filename)
    try:
        crd_dict = get_crds(PATH+filename)
    except:
        print('failed to get crds from file '+filename)
        continue
        
    compare(crd_dict)

        
```
